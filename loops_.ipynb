```python
# Welcome to Hotel
print("1.Idly ---> 50")
print("2.Dosa ---> 80")
print("3.Puri ---> 100")
print("4.Vada ---> 90")
final_bill=0
ch='y'
while ch=='y':

         choice=int(input("Enter your choice : "))
         if choice==1:
            item="Idly"
            cost=50
            print("Idly cost is : ",cost)
         elif choice==2:
              item="Dosa"
              cost=80
              print("Dosa cost is : ",cost)
         elif choice==3:
              item="Puri"
              cost=100
              print("Puri cost is : ",cost)
         else:
              item="Vada"
              cost=90
              print("Vada cost is : ",cost)
         qty=int(input("Enter quantity : "))
         bill_amt=qty*cost
         print("Bill_amt is : ",bill_amt)
         ch=input("Do you want to continue (y\n) : ")
         final_bill +=bill_amt
gst=int(input("Enter gst rate : "))
gst_amt=bill_amt*gst/100
final_bill +=gst_amt
print("Final Bill Amount is : ",final_bill)
```
