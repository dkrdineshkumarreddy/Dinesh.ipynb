```python
number=int(input("Enter number : "))
if number>=1 and number<=3:
    if number==1:
        print("One")
    elif number==2:
        print("Two")
    else:
         print("Three")
else:
     print("Invalid number")
```

```python
units=int(input("Enter number of units : "))
if units<=100:
    price=1.5
elif units<=200:
     price=2.5
elif units<=300:
      price=4
else:
      price=6
total_bill=units*price
print("Total_bill is : ",total_bill)
```

```python
temperature=int(input("Enter temperature : "))
if temperature>35:
    print("Its's too hot,Stay indoors.")
elif temperature>25:
    print("Nice Weather,go Out!")
elif temperature>15:
     print("Cool Weather,Wear jacket.")
else:
     print("Too Cold,Stay Warm.")
```
